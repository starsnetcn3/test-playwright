```python
import os
```

```python
# Variables
IMAGE_DIR = '../output/'
CUSTOMER_URL = 'https:||customer.test.starsnet.hk'
ADMIN_URL = 'https:||admin.test.starsnet.hk'
BACKEND_URL = 'https:||backend.test.starsnet.hk'
```

```python
allImagesPath = os.listdir(IMAGE_DIR)
```

```python
imagesPath = list(filter(lambda path: 'about:blank' not in path, allImagesPath))
```

```python
onlineFlowPath = sorted(list(filter(lambda path: 'online-flow|' in path, imagesPath)))
offlineFlowPath = sorted(list(filter(lambda path: 'offline-flow|' in path, imagesPath)))
```

```python
onlineFlowPath
```

```python
imagesPathWithoutflow = sorted(list(set(imagesPath) - set(onlineFlowPath) - set(offlineFlowPath)))
```

```python
# Static Pages
customerStaticPageImages = list(filter(lambda path: CUSTOMER_URL in path ,imagesPathWithoutflow))
adminStaticPageImages = list(filter(lambda path: ADMIN_URL in path ,imagesPathWithoutflow))

# Api
customerApiImages = list(filter(lambda path: f"{BACKEND_URL}|api|customer" in path,imagesPathWithoutflow))
adminApiImages = list(filter(lambda path: f"{BACKEND_URL}|api|admin" in path,imagesPathWithoutflow))
```
